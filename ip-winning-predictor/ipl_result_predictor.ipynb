```python
 teams = [
'Sunrisers Hyderabad',
'Mumbai Indians',
'Royal Challengers Bangalore',
'Kolkata Knight Riders',
'Kings XI Punjab',
'Chennai Super Kings',
'Rajasthan Royals',
'Delhi Capitals'

 ]
```

```python
match_df['team1']=match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2']=match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1']=match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2']=match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df=match_df[match_df['team1'].isin(teams)]
match_df=match_df[match_df['team2'].isin(teams)]
```

### we will take only those matches in consideration where Duckworth–Lewis is not applied

```python
match_df['dl_applied'].value_counts()
```

```python
match_df=match_df[match_df['dl_applied']==0]
```

```python
match_df.head()
```

```python
delivery.head()
```

```python
match_df=match_df[['match_id','city','winner','total_runs']]
```

```python
match_df.head()
```

```python
delivery_df=match_df.merge(delivery,on='match_id')
delivery_df.head()
```

```python
delivery_df=delivery_df[delivery_df['inning']==2]
delivery_df.head()
```

```python
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df
```

```python
delivery_df['run_left']=delivery_df['total_runs_x']-delivery_df['current_score']
```

```python
delivery_df.head()
```

```python
delivery_df['ball_left']=126-(delivery_df['over']*6+delivery_df['ball'])
```

```python
delivery_df.head()
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna(0)

# Step 2: Convert player_dismissed to binary representation (0 or 1)
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x: 0 if x == 0 else 1)

# Step 3: Ensure the column is of integer type
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype(int)

# Step 4: Calculate cumulative wickets lost per match
delivery_df['player_dismissed'] = delivery_df.groupby('match_id')['player_dismissed'].cumsum()

# Step 5: Calculate remaining wickets
delivery_df['wickets'] = 10 - delivery_df['player_dismissed']
delivery_df.head()

```

```python
delivery_df.tail()
```

```python
delivery_df['crr']=(delivery_df['current_score']*6)/(120-delivery_df['ball_left'])
```

```python
delivery_df.head()
```

```python
delivery_df['rrr']=(delivery_df['run_left']*6)/(delivery_df['ball_left'])
```

```python
delivery_df.tail()
```

```python
def result(row):
    return 1 if row['batting_team']==row['winner'] else 0
```

```python
delivery_df['result']=delivery_df.apply(result,axis=1)
```

```python
delivery_df.tail()
```

```python
final_df=delivery_df[['batting_team','bowling_team','city','run_left','ball_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df
```

```python
final_df=final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df=final_df[final_df['ball_left']!=0]
```

```python

    X = final_df.iloc[:, :- 1]
    y = final_df.iloc[:,-1]
    from sklearn.model_selection import train_test_split
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
('trf', OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline(steps=[
    ('steps1',trf),
    ('steps2',LogisticRegression(solver='liblinear')),
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
